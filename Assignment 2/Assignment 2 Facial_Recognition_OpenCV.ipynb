## 2. Import Required Libraries

```python
import cv2
import matplotlib.pyplot as plt

# Helper function to display images in Jupyter
def show_image(img_bgr, title="Image"):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

## 3. Load Haar Cascade Classifier

```python
# Load pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

if face_cascade.empty():
    print("Error loading Haar Cascade XML file") 
else:
    print("Haar Cascade classifier loaded successfully")
```

## 4. Detect Faces in an Image

```python
# Load an image
image_path = r"C:\Users\Admin\OneDrive\Desktop\DL assignments\Assignment 2\Saved Pictures\283A6637.JPG"  # Replace with your image path
img = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Show result
show_image(img, title="Detected Faces")
print(f"Found {len(faces)} face(s) in the image.")
```

## 5. Real-Time Face Detection using Webcam

```python
## 5. Real-Time Face Detection using Webcam (Jupyter Output)

import cv2
import matplotlib.pyplot as plt

# Open webcam
cap = cv2.VideoCapture(0)

# Capture a single frame
ret, frame = cap.read()
cap.release()

if ret:
    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Draw rectangles
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Convert BGR → RGB for matplotlib
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Show result in notebook output
    plt.figure(figsize=(8,6))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(f"Detected {len(faces)} face(s)")
    plt.show()
else:
    print("Failed to capture frame from webcam")
```
